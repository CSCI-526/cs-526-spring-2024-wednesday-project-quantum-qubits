```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Level': ['Level 1', 'Level 2', 'Level 3'],
    'Average Number of Jumps Made by the Player Before Winning by Level': [4.5,3.5,2.5]
})

df.plot(x="Level", y=["Average Number of Jumps Made by the Player Before Winning by Level"], kind="bar", title="Average Number of Jumps Made by the Player Before Winning by Level")
plt.xlabel("Level Number")
plt.ylabel('Average Number of Jumps Made by the Player Before Winning by Level')
plt.tight_layout()
plt.show()
```
